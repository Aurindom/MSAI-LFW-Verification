# LFW Face Verification — Exploration
Load processed data and inspect dataset statistics.

```python
import sys, json
import numpy as np
import matplotlib.pyplot as plt
sys.path.insert(0, '..')
from src.ingest import load_config

cfg = load_config()
out_dir = cfg['data']['out_dir']

images = np.load(f'{out_dir}/images.npy')
labels = np.load(f'{out_dir}/labels.npy')
names  = np.load(f'{out_dir}/names.npy')

with open(f'{out_dir}/manifest.json') as f:
    manifest = json.load(f)

print(json.dumps(manifest, indent=2))
```

```python
unique, counts = np.unique(labels, return_counts=True)
fig, axes = plt.subplots(1, min(8, len(unique)), figsize=(16, 3))
for ax, uid in zip(axes, unique[:8]):
    idx = np.where(labels == uid)[0][0]
    ax.imshow(images[idx], cmap='gray')
    ax.set_title(names[uid], fontsize=7)
    ax.axis('off')
plt.tight_layout()
plt.show()
```
